```r
library(ggplot2)
```

```r
data=read.csv("out_mr_cho.csv")
```

```r
df=as.data.frame(data)
```

```r
df <- subset(df, select = -c(X) )
df$X2 <- gsub("b" ,"", df$X2)
df
```

```r
png('Kmeans_Plot_MR.jpg')
ggplot(df, aes(y=df$X1, x=df$X0, color=df$X2)) + scale_x_continuous("Principal_Component1") +
  scale_y_continuous("Principal_Component2") +
  ggtitle("PCA Plot")+labs(colour="Cluster")+
    geom_point(size=2, alpha=0.6)
dev.off()
```
